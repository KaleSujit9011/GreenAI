```python
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}
```

```python
ec_m = df['Energy Consumption (MWh)'].mean()
print("Mean of Energy Consumption (MWh): ",ec_m)
```

```python
c_m = df['Cost (Million $)'].mean()
print("Mean of Cost (Million $): ",c_m)
```

```python
df.isnull()
```

## outlier - 3*standard deviation
### ffill(forward fill) and bfill(backward)

```python
df.ffill()
```

```python
df.bfill()
```

```python
df['Energy Consumption (MWh)'] = df['Energy Consumption (MWh)'].fillna(ec_m)
```

```python
df['Cost (Million $)'] = df['Cost (Million $)'].fillna(c_m)
```

```python
df
```

#### 1)normalization 
#### (bet 0-1)minmaxscaler = (xi-xmin)/(xmax - xmin) 


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# energy_df = pd.DataFrame(data)
# from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df[["Energy Consumption (MWh)","Cost (Million $)"]] = scaler.fit_transform(df[["Energy Consumption (MWh)", "Cost (Million $)"]])
df.head()
```

#### 2)(meu-mean=0,std-sigma=1)standardization/Z-score scalar = xi -meu/sigma

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler  = StandardScaler()
```

```python
df[["Energy Consumption (MWh)","Cost (Million $)"]] = scaler.fit_transform(df[["Energy Consumption (MWh)", "Cost (Million $)"]])
df.head()
```
