```python
# if __init__ == "__main__":

def total_consumption(energy_consumption,n,total):
    if total < 1:
        total = 0
    total +=  energy_consumption
    return total

n = int(input("no. of building= "))
for i in range(n):
    energy_consumption = int(input(f"EC{i+1}: "))

    if energy_consumption > 1000:
        print("Energy Intensive use.")
    elif energy_consumption <1000 and energy_consumption>500:
        print("Moderate Consumption.")
    else :
        print("Energy Efficient")

    if energy_consumption > 10000:
        print("consumption exceeds a critical grid limit")
        break
        
    total = total_consumption(energy_consumption,n,0)

avg = lambda total:total/n
print("Total: ",total)
print("avg: ",avg)
```
